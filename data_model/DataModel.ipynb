```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Graph

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)

# Explicitly set positions
pos = {1:(0,0), 2:(1,0), 3:(0,1)}

options = {
    'font_size': 12,
    'node_size': 3000,
    'node_color': 'white',
    'edgecolors': 'black',
    'linewidths': 3,
    'width': 3
}

nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis('off')
plt.show()
```

## Weighted Graph

```python
G = nx.Graph()
G.add_edge(1, 2, weight=1)
G.add_edge(1, 3, weight=0.5)

# Explicitly set positions
pos = {1:(0,0), 2:(1,0), 3:(0,1)}

options = {
    'font_size': 12,
    'node_size': 3000,
    'node_color': 'white',
    'edgecolors': 'black',
    'linewidths': 3,
    'width': 3
}

nx.draw_networkx(G, pos, **options)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis('off')
plt.show()
```

## Directed Graph

```python
G = nx.DiGraph()
G.add_edge(1, 2, weight=1)
G.add_edge(1, 3, weight=0.5)

# Explicitly set positions
pos = {1:(0,0), 2:(1,0), 3:(0,1)}

options = {
    'font_size': 12,
    'node_size': 3000,
    'node_color': 'white',
    'edgecolors': 'black',
    'linewidths': 3,
    'width': 3
}

nx.draw_networkx(G, pos, **options)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis('off')
plt.show()
```
