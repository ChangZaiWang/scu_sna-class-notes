```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
df = pd.read_csv("./archive/bs140513_032310.csv")
df.head()
```

```python
# split the features and labels
label = df.fraud
features = df.drop('fraud', axis = 1)
features.head()
```

```python
# check if empty values...
df.isnull().values.any()
```

```python
features =  features.drop(['step', 'zipcodeOri', 'zipMerchant'], axis = 1)
features.head()
```

```python
features_dum = pd.get_dummies(features[['age', 'category', 'gender']])
features = pd.concat([features_dum,features],axis=1)
```

```python
features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler, then apply it to the features
scaler = MinMaxScaler()

features[[ 'amount' ]]=scaler.fit_transform(features[[ 'amount' ]])

# Show an example of a record with scaling applied
features.head()
```

```python
features =  features.drop(['customer', 'merchant', 'age', 'gender', 'category'], axis = 1)
features.head()
```

```python
features.shape
```

```python
from sklearn.model_selection import * 
from sklearn.metrics import fbeta_score, accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.33, random_state=42)
```

```python
clfRandomForest = RandomForestClassifier(n_jobs=8)
clfRandomForest.fit(X_train,y_train)
    
predictionsRF = clfRandomForest.predict(X_test)
    
print(fbeta_score(y_test, predictionsRF, average='macro', beta=1))
print(accuracy_score(y_test, predictionsRF))
print(roc_auc_score(y_test, predictionsRF))
```

## Network feature

```python
features_graph = pd.read_csv("./archive/bs140513_032310.csv")
features_graph.head()
```

```python
# split the features and labels
label = features_graph.fraud
features_graph = features_graph.drop('fraud', axis = 1)
features_graph.head()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Build your graph
G = nx.from_pandas_edgelist(features_graph,'customer', 'merchant','amount', create_using=nx.DiGraph() )
```

```python
r = nx.pagerank(G, alpha=0.9)
```

```python
features_graph['custPR']= features_graph['customer'].map(r)
```

```python
features_graph['merPR']= features_graph['merchant'].map(r)
```

```python
features_graph.head()
```

```python
h,a=nx.hits(G)
```

```python
features_graph['custHITS_h']= features_graph['customer'].map(h)
```

```python
features_graph['custHITS_a']= features_graph['customer'].map(a)
```

```python
features_graph['merHITS_h']= features_graph['merchant'].map(h)
```

```python
features_graph['merHITS_a']= features_graph['merchant'].map(a)
```

```python
features_graph.head()
```

```python
d = G.degree()
```

```python
features_graph['custdegree']= features_graph['customer'].map(d)
```

```python
features_graph['merdegree']= features_graph['merchant'].map(d)
```

```python
features_graph.head()
```

```python
import networkx as nx
cn = nx.closeness_centrality(G)
```

```python
features_graph['custcloseness']= features_graph['customer'].map(cn)
```

```python
features_graph['mercloseness']= features_graph['merchant'].map(cn)
```

```python
features_graph.head()
```

```python
bt = nx.betweenness_centrality(G)
```

```python
features_graph['custbetweenness']= features_graph['customer'].map(bt)
```

```python
features_graph['merbetweenness']= features_graph['merchant'].map(bt)
```

```python
features_graph.head()
```

```python
features_graph.columns
```

```python
features_graph.describe()
```

```python
scaler = MinMaxScaler()

features_graph[['amount', 'custPR', 'merPR', 'custHITS_h',
       'custHITS_a', 'merHITS_h', 'merHITS_a', 'custdegree', 'merdegree',
       'custcloseness', 'mercloseness', 'custbetweenness', 'merbetweenness']] = scaler.fit_transform(features_graph[['amount' , 'custPR', 'merPR', 'custHITS_h',
       'custHITS_a', 'merHITS_h', 'merHITS_a', 'custdegree', 'merdegree',
       'custcloseness', 'mercloseness', 'custbetweenness', 'merbetweenness']])

# Show an example of a record with scaling applied
features_graph.head()
```

```python
features_graph =  features_graph.drop(['step', 'customer', 'merchant'], axis = 1)
features_enhanced = pd.get_dummies(features_graph)
```

```python
features_enhanced.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_enhanced, label, test_size=0.33, random_state=42)
```

```python
clfRandomForest = RandomForestClassifier(n_jobs=8)
clfRandomForest.fit(X_train,y_train)
    
predictionsRF = clfRandomForest.predict(X_test)
    
print(fbeta_score(y_test, predictionsRF, average='macro', beta=1))
print(accuracy_score(y_test, predictionsRF))
print(roc_auc_score(y_test, predictionsRF))
```
